### Generate Metadata

Generate a CSV file with 10,000 accounts

Parameters:

 - t 1 : use 1 thread
 - m metadata : generate customer metadata (vs only account ID)
 - d 1 : only generate 1 month (useful for statements)

```python
!rm metadata.csv
!../scripts/gen.sh -t 1 -m metadata -n 10000 -d 1
```

### Generate ID cards

```python
!rm metadata.csv
! mkdir -p idcards
! rm idcards/*
!../scripts/gen.sh -t 1 -m pdf -d 1 -n 10 -f jpeg  -x id -o file:idcards
```

```python
!ls idcards
```

```python
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
##%matplotlib inline

images = []
for img_path in glob.glob('idcards/*.jpg'):
    images.append(mpimg.imread(img_path))

plt.figure(figsize=(15,10))
columns = 3
for i, image in enumerate(images):
    plt.subplot(int(len(images) / columns) + 1, columns, i + 1)
    plt.imshow(image)
    plt.axis('off')
```
