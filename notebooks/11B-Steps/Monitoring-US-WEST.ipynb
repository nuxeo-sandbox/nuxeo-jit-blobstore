```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

cpu = pd.read_csv("monitoring/US-WEST-IMPORT-CSV/extract-2020-09-14T20_24_38.673Z-avg_system.cpu.idle{staging_bench-11} by {hostname}.csv")
commits = pd.read_csv("monitoring/US-WEST-IMPORT-CSV/extract-2020-09-14T20_25_14.610Z-per_second(sum_nuxeo.importer.stream.consumer.committed{staging_bench-11}).csv");
iowait = pd.read_csv("monitoring/US-WEST-IMPORT-CSV/extract-2020-09-14T20_25_29.070Z-max_mongodb.atlas.system.cpu.norm.iowait{projectname_nxbench-bench11}.csv");


cpuLines=cpu.group.str.contains("10-3-10-10")
importerCpu=cpu[cpuLines]
importerCpu['cpu'] = importerCpu.apply(lambda row: 100-row.value, axis = 1) 


fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add CPU
fig.add_trace(
    go.Scatter(x = importerCpu['time'], y = importerCpu['cpu'],name='Importer CPU(%)'),
    secondary_y=False,
)

# Add IOWait
fig.add_trace(
    go.Scatter(x = iowait['time'], y = iowait['value'],name='MongoDB IOWait(%)'),
    secondary_y=False,
)

# Add throuput
fig.add_trace(
    go.Scatter(x=commits['time'], y=commits['value'], line= {"color":"green"}, name="Documents imported/s", mode='lines'),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="Import 800M on MongoDb Atlas M60 NVMe - Speed vs IOWait"
)

# Set x-axis title
fig.update_xaxes(title_text="Time")

# Set y-axes titles
fig.update_yaxes(title_text="<b>CPU %</b>", secondary_y=False, range=[0,100])
fig.update_yaxes(title_text="<b>Throughtput</b> doc/s", secondary_y=True)

fig.show()

```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

commits = pd.read_csv("monitoring/US-WEST-IMPORT-CSV/extract-2020-09-14T20_25_14.610Z-per_second(sum_nuxeo.importer.stream.consumer.committed{staging_bench-11}).csv");
writeLatency = pd.read_csv("monitoring/US-WEST-IMPORT-CSV/extract-2020-09-14T20_25_33.305Z-max_mongodb.atlas.oplatencies.writes.avg{projectname_nxbench-bench11}.csv")


fig = make_subplots(specs=[[{"secondary_y": True}]])

# MongoDB Latency
fig.add_trace(
    go.Scatter(x = writeLatency['time'], y = writeLatency['value'],name='MongoDB Avg Write Latency'),
    secondary_y=False,
)

# Add throuput
fig.add_trace(
    go.Scatter(x=commits['time'], y=commits['value'], line= {"color":"green"}, name="Documents imported/s", mode='lines'),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="Import 800M on MongoDb Atlas M60 NVMe  - Speed vs Write Latency"
)

# Set x-axis title
fig.update_xaxes(title_text="Time")

# Set y-axes titles
fig.update_yaxes(title_text="<b>Time</b> ms", secondary_y=False)
fig.update_yaxes(title_text="<b>Throughtput</b> doc/s", secondary_y=True)

fig.show()

```
